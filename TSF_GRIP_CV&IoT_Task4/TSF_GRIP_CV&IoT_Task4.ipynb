### 1.1 Importing required modules

```python
import cv2
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import numpy as np
```

### 1.2 Associating the required xml file

```python
harcs_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

### 1.3 Acquiring data of images with no mask

```python
capture = cv2.VideoCapture(0)
data_nomask=[]
while True:
    flag, image = (capture.read())
    if flag:
        faces = harcs_data.detectMultiScale(image) 
        for x,y,w,h in faces:
            cv2.rectangle(image, (x,y), (x+w, y+h), (200,125,255), 4)
            face = image[y:y+h, x:x+h, :]
            face = cv2.resize(face, (100,100))
            print(len(data_nomask))
            if len(data_nomask) < 400:
                data_nomask.append(face)
                
        cv2.imshow('Cam', image)
        if cv2.waitKey(2)==27 or len(data_nomask) > 200:
            break
                
capture.release()
cv2.destroyAllWindows()
            
            
```

```python
np.save('without_mask.npy', data_nomask)
```

### 1.4 Acquiring data of images with mask

```python
cap = cv2.VideoCapture(0)
data_mask=[]
while True:
    flag, image = (cap.read())
    if flag:
        faces = harcs_data.detectMultiScale(image) 
        for x,y,w,h in faces:
            cv2.rectangle(image, (x,y), (x+w, y+h), (200,125,255), 4)
            face = image[y:y+h, x:x+h, :]
            face = cv2.resize(face, (100,100))
            print(len(data_mask))
            if len(data_mask) < 400:
                data_mask.append(face)
            cv2.imshow('Cam', image)    
        if cv2.waitKey(2)==27 or len(data_mask) > 200:
            break
                
cap.release()
cv2.destroyAllWindows()
```

```python
np.save('mask.npy', data_mask)
```

### 1.5 Processing the data

```python
no_mask = np.load('without_mask.npy')
mask = np.load('mask.npy')
```

```python
mask.shape
```

```python
no_mask.shape
```

```python
mask = mask.reshape(201, 100*100*3)
```

```python
no_mask = no_mask.reshape(201, 100*100*3)
```

```python
X = np.r_[no_mask, mask]
```

```python
labels = np.zeros(X.shape[0])
labels[201:] = 1.0
```

### 1.6 Preparing the model

```python
x_train, x_test, y_train, y_test = train_test_split(X, labels, test_size=0.25)
```

```python

pca =PCA(n_components = 3)
x_train = pca.fit_transform(x_train)

svm= SVC()
svm.fit(x_train, y_train)
```

```python
x_test = pca.transform(x_test)
y_pred = svm.predict(x_test)
```

```python
name={0:'No mask', 1: 'Mask'}
```

### 1.7 Detection using webcam

```python
capture = cv2.VideoCapture(0)


font_scale = 2
font = cv2.FONT_HERSHEY_DUPLEX

while True:
    flag, image = (capture.read())
    if flag:
        faces = harcs_data.detectMultiScale(image) 
        for x,y,w,h in faces:
            cv2.rectangle(image, (x,y), (x+w, y+h), (200,125,255), 4)
            face = image[y:y+h, x:x+h, :]
            face = cv2.resize(face, (100,100))
            face = face.reshape(1,-1)
            face = pca.transform(face)
            pred = svm.predict(face)
            cv2.putText(image, name[int(pred)],(100,100), font, fontScale=font_scale, color=(0,240,0))
            
                
        cv2.imshow('Cam', image)
        if cv2.waitKey(2)==27 :
            break
                
capture.release()
cv2.destroyAllWindows()
```
